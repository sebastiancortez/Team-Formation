## Diversity Index of Each Group

```python
# A group can be a herd or the whole population
def calculate_SDI(group):
    species = group.groupby('Specialization').count()
    species['n-1'] = species['Respondent #'] - 1
    species['n(n-1)'] = species['Respondent #']*species['n-1']
    nips = species['Respondent #'].sum()
    tni = species['n(n-1)'].sum()
    #print(nips, tni)
    SDI = 1- (tni/(nips*(nips-1)))
    return SDI
```

## Mutation algorithm

The algorithm below takes a population dataframe a swap rows randomly and then form the herds

```python
def mutate(population):
# new_population = population
    end_index = len(population.index)-1
    first_random = random.randint(0, end_index)
    second_random = random.randint(0, end_index)
    def half_len_round(x,base):
        return base * math.ceil((x/2)/base)
    group_number1 = 7 
    mid_index = half_len_round(len(population.index),group_number1)-1
    #allocating those with client priority to their preferred projects
    ##need to allocate the remaining based on preferences?
    ##clean up to just reflect 'project 1' and 'project 2' rather than project names - can clean up code after (redundant)
    if (str(population.iloc[first_random,3]) == 'Priority Preference'):
        if population.iloc[first_random,2] == 'Remote Community Health Services':
            if first_random > mid_index:
                second_random = random.randint(0, mid_index)
            else:
                second_random = first_random
        elif population.iloc[first_random,2] == 'Sustainable Use of Fishery Resources':
            if first_random <= mid_index:
                second_random = random.randint(mid_index+1,end_index)
            else:
                second_random = first_random
    elif (str(population.iloc[second_random,3]) == 'Priority Preference'):
        if population.iloc[second_random,2] == 'Remote Community Health Services':
            if second_random > mid_index:
                first_random = random.randint(0, mid_index)
            else:
                first_random = second_random
        elif population.iloc[second_random,2] == 'Sustainable Use of Fishery Resources':
            if second_random <= mid_index:
                first_random = random.randint(mid_index+1,end_index)
            else:
                first_random = second_random
    individual_a, individual_b = population.iloc[first_random].copy(), population.iloc[second_random].copy()
    population.iloc[first_random], population.iloc[second_random] = individual_b, individual_a
    return population
```
