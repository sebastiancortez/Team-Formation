# Group Creation

```python
def herd_formation(population):
    for i in range(0,len(population.index),6):
        x = int(i/6)
        exec('herd{} = pd.DataFrame()'.format(x))
        globals()['herd{}'.format(x)] = population.iloc[i:i+6:, :]
```

## Diversity Index of Each Group

```python
# A group can be a herd or the whole population
def calculate_SDI(group):
    species = group.groupby('Specialization').count()
    species['n-1'] = species['Respondent #'] - 1
    species['n(n-1)'] = species['Respondent #']*species['n-1']
    nips = species['Respondent #'].sum()
    tni = species['n(n-1)'].sum()
    #print(nips, tni)
    SDI = 1- (tni/(nips*(nips-1)))
    return SDI
```

## Mutation algorithm

The algorithm below takes a population dataframe a swap rows randomly and then form the herds

```python
def mutate(population):
#    new_population = population
    first_random = random.randint(0, 46)
    second_random = random.randint(0, 46)
    individual_a, individual_b = population.iloc[first_random].copy(), population.iloc[second_random].copy()
    population.iloc[first_random], population.iloc[second_random] = individual_b, individual_a
    return population
```

## Evaluation algorithm

```python
# In this case the average SDI and the standard deviation will be the values to evaluate the fitness of the population
def average_SDI(population):
    herd_formation(population)
    y = []
    for i in range(0,int(len(df_sample1.index)/6)):
        x = calculate_SDI(globals()['herd{}'.format(i)])
        y.append(x)
        print(x)
    a_SDI = mean(y)
    std_SDI = stdev(y)
    return (a_SDI, std_SDI)
```
